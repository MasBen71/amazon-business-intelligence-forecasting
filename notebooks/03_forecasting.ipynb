# 03 — Forecasting

This notebook trains a compact revenue forecasting model and generates segment revenue forecasts for 2026-2028.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(ROOT))

from src.data_loader import load_raw_data
from src.features import add_segment_features
from src.modeling import forecast_revenue

raw = load_raw_data(ROOT / "data/raw/amazon_segment_annual_2019_2025.csv")
features = add_segment_features(raw)
result = forecast_revenue(features)
result.metrics
```

```python
result.forecast
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
for segment, group in features.groupby("segment"):
    ax.plot(group["year"], group["net_sales_musd"] / 1000, marker="o", label=f"{segment} actual")
for segment, group in result.forecast.groupby("segment"):
    ax.plot(group["year"], group["forecast_net_sales_musd"] / 1000, marker="x", linestyle="--", label=f"{segment} forecast")
ax.set_title("Amazon Segment Revenue Forecast")
ax.set_xlabel("Fiscal year")
ax.set_ylabel("USD billions")
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)
plt.show()
```
