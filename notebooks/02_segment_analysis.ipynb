# 02 — Segment Analysis

This notebook creates business KPIs and visualizes revenue, profit, and margin differences across Amazon's segments.

```python
from pathlib import Path
import sys
import matplotlib.pyplot as plt

ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(ROOT))

from src.data_loader import load_raw_data
from src.features import add_segment_features, latest_year_summary

raw = load_raw_data(ROOT / "data/raw/amazon_segment_annual_2019_2025.csv")
features = add_segment_features(raw)
features.head()
```

```python
latest_year_summary(features)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
for segment, group in features.groupby("segment"):
    ax.plot(group["year"], group["net_sales_musd"] / 1000, marker="o", label=segment)
ax.set_title("Amazon Segment Revenue")
ax.set_xlabel("Fiscal year")
ax.set_ylabel("USD billions")
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
for segment, group in features.groupby("segment"):
    ax.plot(group["year"], group["operating_margin"] * 100, marker="o", label=segment)
ax.axhline(0, linewidth=1)
ax.set_title("Amazon Segment Operating Margin")
ax.set_xlabel("Fiscal year")
ax.set_ylabel("Operating margin, %")
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```
